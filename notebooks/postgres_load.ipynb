#### Some SQL Queries

```python
# Create a connection string for SQLAlchemy to connect to PostgreSQL

connection_string = (
    f"postgresql+psycopg2://{DB_USER}:{DB_PASSWORD}@{DB_HOST}:{DB_PORT}/{DB_NAME}"
)
```

```python
# Call a function to execute predefined SQL queries
results = execute_telecom_queries(connection_string)

# Print each result separately with clear formatting
for key, value in results.items():
    print(f"{key}:\n")
    
    # Check if the result is a DataFrame
    if isinstance(value, pd.DataFrame):
        print(value.to_string(index=False))  # Convert DataFrame to string without index
    else:
        print(value)  # Print other types directly
    
    print("\n" + "="*40 + "\n")  # Separator for clarity
```

```python
# Create a histogram to visualize the distribution of call durations

plt.figure(figsize=(10, 6))  # Set the figure size
sns.histplot(
    df["Dur. (ms)"], bins=30, kde=True, color="green"
)  # Plot histogram with Kernel Density Estimate (KDE)
plt.title("Distribution of Call Duration")  # Title of the plot
plt.xlabel("Duration (ms)")  # X-axis label
plt.ylabel("Frequency")  # Y-axis label
plt.show()  # Display the plot
```

```python
# Calculate the average duration of calls grouped by the last location name

avg_duration_by_location = (
    df.groupby("Last Location Name")["Dur. (ms)"].mean().reset_index()
)

# Sort the results by average duration in descending order and print them
print(avg_duration_by_location.sort_values(by="Dur. (ms)", ascending=False))
```
